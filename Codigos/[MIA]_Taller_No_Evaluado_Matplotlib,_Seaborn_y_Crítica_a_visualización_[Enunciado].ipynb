## Actividad 5 - Similitudes de Lenguajes


El 2012 se publicó la siguiente visualización para observar las similitudes de los lenguajes en el mundo. En particular, para permitir al usuario identificar cuales lenguajes son similares según una métrica que mide diferencia de palabras (distancia de _Levenshtein_).


![AC06](https://res.cloudinary.com/hernan4444/image/upload/v1684177382/T5_2023_1_Caso2.jpg)

Fuente: https://hernan4444.github.io/Language-Network/ *

\* La página original es [esta](http://languagenetwork.cotrino.com/) pero ahí no funciona la visualización.

En esta visualización interactiva:
- Cada nodo circular representa un lenguaje. Su tamaño es proporcional al uso de dicho lenguaje en los datos utilizados. El color no codifica ninguna información, solo fue utilizado con fines estéticos.
- Cada línea entre dos nodos representa la similitud de los lenguajes.
- La posición de cada nodo no representa ninguna información, esta fue generada por un algoritmo que intentó reducir la cantidad de enlaces que se cruzan.
- El saturación de cada línea codifica el porcentaje de similitud. Una línea más saturada, indica que hay una mayor similaridad.

Respecto a las interacciones posibles:

- Se puede utilizar el _mouse_ para hacer _drag & drop_ en la visualización. Esto permite mover un nodo y volver a aplicar el algoritmo que determine la posición de todos los demás nodos.

- Se puede hacer _hover_ sobre un nodo. Esto permite ver información más detallada de este lenguaje y la similitud con otros lenguajes.

- Inicialmente se muestran todos los nodos del _dataset_, y con un _checkbox_ es posible filtrar aquellos nodos que no presentan enlaces. Permitiendo ver únicamente los lenguajes que tengan conexión con otro lenguaje.

- Inicialmente la visualización filtra aquellas líneas que representan un porcentaje menor al 50% de similitud. Luego, mediante un selector en la barra izquierda, es posible alterar ese filtro en un rango de 33% a 85% de similaridad. Permitiendo aumentar y reducir la cantidad de líneas que se ven en la visualización.
